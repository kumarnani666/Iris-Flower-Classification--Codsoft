```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
data_iris['species']=le.fit_transform(data_iris['species'])
data_iris.head(100)
```

```python
sns.scatterplot(data=data_iris,x="sepal_length",y="sepal_width",hue="species")
```

```python
sns.scatterplot(data_iris,x="petal_length",y="petal_width",hue="species")
```

```python
x=data_iris[['sepal_length','sepal_width','petal_length','petal_width']]
y=data_iris[['species']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

MODEL PREDICTION

```python
x_train_prediction=log.predict(x_train)
print(x_train_prediction)
```

```python
x_test_prediction=log.predict(x_test)
print(x_test_prediction)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.metrics import accuracy_score
training_data_accuracy=accuracy_score(y_train,x_train_prediction)
```

```python
print("Accuracy score of training data: ",training_data_accuracy)
```

```python
testing_data_accuracy=accuracy_score(y_test,x_test_prediction)
print("Accuracy score of testing data: ",testing_data_accuracy)
```

```python
import warnings
warnings.filterwarnings('ignore')

res = log.predict([[5.8, 2.6, 4, 1.2]])

if res == 0:
    print("Iris-setosa")
elif res == 1:
    print("Iris-versicolor")
else:
    print("Iris-virginica")
```
